1. Pel√≠cula con mayor duraci√≥n con filtros opcionales de A√ëO, PLATAFORMA Y TIPO DE DURACI√ìN. (la funci√≥n debe llamarse get_max_duration(year, platform, duration_type))

```python
def get_max_duration(year:int, plataform: str, duration_type:str ):
    
    dt_f = df[(df['release_year'] == year) & (df['plt'] == plataform)]
    
    if duration_type == 'min':
        titulo = dt_f.loc[dt_f.duration_int.idxmax()]['title']
    else:
        titulo = dt_f.loc[dt_f.duration_int.idmax()]['title']
    return titulo
```

```python
def get_score_count(platform: str, scored: int, year: int):
    count = df_rating[df_rating['movieid'].str.contains(platform[0], case = False)]
    cant = count[(count.rating > scored) & (count.release_year == year)]

    return {'plataforma':platform, 'cantidad': len(cant)}
```

3. Cantidad de pel√≠culas por plataforma con filtro de PLATAFORMA. (La funci√≥n debe llamarse get_count_platform(platform))

```python
def get_count_platform(platform:str):
    res_1 = ((df['plt'] == platform) & (df['type'].str.contains('Movie'))).sum()
    res_2 = ((df['plt'] == platform) & (df['type'].str.contains('TV Show'))).sum()

    return ('Plataforma: ' + str(platform) + '  ' + 'Movie: '  + str(res_1)+ '  ' + 'TV Show: '  + str(res_2))
            
```

```python
def get_actor(platform:str, year:int):
    data_actor = df[(df['plt'] == platform) & (df['release_year'] == year)]
    for i in data_actor['cast']:
        if i != 'Sin dato':
            i = i.replace(', ', ',')
        else:
            continue
    lista = []
    for i in data_actor['cast']:
        if i != 'Sin dato':
            n=i.split(',')
            for j in range(len(n)):
                if n[j] not in lista:
                    lista.append(n[j])
                else:
                    continue
        else:
            continue
    
    lista = list(set(lista))
    m = 0
    diccionario = {}
    for i in lista:
        m = 0
        for j in data_actor['cast']:
            if i in j.split(','):
                m += 1
        diccionario[i] = m
    return print(max(diccionario, key=diccionario.get))
    
```
